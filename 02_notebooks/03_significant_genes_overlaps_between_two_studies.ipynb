```r
here::here()
```

```r
library(pacman)
p_load("DESeq2", "pheatmap", "EnhancedVolcano", "biomaRt", "GEOquery", "dplyr", "GENIE3", "here", install = FALSE)
```

```r
dataset_1 <- read.csv(here("03_outputs", "GSE89408_correlation_matrix.csv"), row.names = 1)
dataset_2 <- read.csv(here("03_outputs", "E-MTAB-6141_correlation_matrix.csv"), row.names = 1)
```

```r
dim(dataset_1)
dim(dataset_2)
```

```r
intersected_genes_rows <- intersect(rownames(dataset_1), rownames(dataset_2))
intersected_genes_cols <- intersect(names(dataset_1), names(dataset_2))
```

```r
dataset_1 <- dataset_1[intersected_genes_rows,intersected_genes_cols] %>% as.matrix()
dataset_2 <- dataset_2[intersected_genes_rows,intersected_genes_cols] %>% as.matrix()
```

```r
average_matrix <- (dataset_1 + dataset_2) / 2
```

```r
abs_cor <- abs(average_matrix)
threshold <- quantile(abs_cor, probs = 0.75, na.rm = TRUE)
filtered_cor_matrix <- ifelse(abs_cor >= threshold, average_matrix, 0)
filtered_cor_matrix
```

```r
write.csv(dataset_1, here("03_outputs", "GSE89408_correlation_mtx_common_genes.csv", quote = FALSE))
```

```r
write.csv(dataset_2, here("03_outputs", "E-MTAB-6141_correlation_mtx_common_genes.csv", quote = FALSE))
```

```r
write.csv(filtered_cor_matrix, here("03_outputs", "correlation_mtx_merged_filtered.csv"))
```
